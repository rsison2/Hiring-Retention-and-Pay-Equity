```python
#Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import HR data csv as dataframe
df = pd.read_excel("H:/2305-FTB-CT-DAT-PT/Capstone/HRDataset_v14(v2).xlsx")
```

```python
#Check to confirm dataframe created... preliminary grid view of dataframe
#Upload of csv and creation of dataframe visually confirmed
display(df)
```

```python
#Convert TermReason field to proper case... field has lowercase letters as first letter
df['TermReason'] = df['TermReason'].str.title()

display(df['TermReason'])
```

```python
#Check column names, non-null count, and data types
#All columns have all non-null values except for DateofTermination
df.info()
```

```python
#Convert date fields from object to datetime
df['DOB'] = pd.to_datetime(df['DOB'])
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DateofTermination'] = pd.to_datetime(df['DateofTermination'])
df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])
```

```python
#Chcek for null values by creating a new dataframe

pd.DataFrame(
    df.isnull().sum(),
    columns=["Missing Values"]
)
```

```python
#Check which ManagerNames are associated with null values in ManagerID column
#All null values in ManagerID column are associated with Webster Butler
managerid_nulls = df[df['ManagerID'].isnull()]
manager_names = managerid_nulls['ManagerName']

print(manager_names)
```

```python
#Check to see which ManagerIDs are associated with Webster Butler within the dataset as a whole
#39.0 is the ManagerID that is associated with Webster Butler
managerNameCheck = df[df['ManagerName'] == 'Webster Butler']
manager_ids = managerNameCheck['ManagerID']

print(manager_ids)
```

```python
#Replace the null values in ManagerID field with correct ID for Webster Butler; Butler is the only
# manager for whom there are null entries in the ManagerID field

df['ManagerID'] = df['ManagerID'].fillna(39.0)
```

```python
#Chcek for null values again... there are no more null values in ManagerID
pd.DataFrame(
    df.isnull().sum(),
    columns=["Missing Values"]
)
```

```python
#Show duplicate values and count of dup values in each column of dataframe by
#using a for loop

for column in df.columns:
    duplicates = df[column][df.duplicated(subset=column, keep=False)]
    counts = duplicates.value_counts()
    print(f"Column: {column}")
    print(counts)
```

```python
#Show list of employee names associated with duplicate DOB values in datafram
#using a for loop
#Indicates if there is duplicate employees in dataset because of name mispelling

duplicate_birthdays = df[df.duplicated(subset='DOB', keep=False)]

for date_of_birth, group in duplicate_birthdays.groupby('DOB'):
    print(f"DOB: {date_of_birth}")
    for _, row in group.iterrows():
        print(f"Employee_Name: {row['Employee_Name']}")
    print()
```

```python
#Check all columns in dataframe for cells with leading or trailing spaces

for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, str)).any():
        has_spaces = df[col].apply(lambda x: isinstance(x, str) and (x != x.lstrip() or x != x.rstrip()))
        if has_spaces.any():
            print(f"Column '{col}' has leading or trailing spaces.")
            print(df.loc[has_spaces, col])
        else:
            print(f"Column '{col}' has no leading or trailing spaces.")
    else:
        print(f"Column '{col}' is not of string type, so no leading or trailing spaces to check.")
```

```python
#Remove leading and trailing spaces from all columns in the dataframe of object datatype
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip()
```

```python
#Check all columns again for cells with leading or trailing spaces... all leading and trailing spaces have been
#removed

for col in df.columns:
    if df[col].apply(lambda x: isinstance(x, str)).any():
        has_spaces = df[col].apply(lambda x: isinstance(x, str) and (x != x.lstrip() or x != x.rstrip()))
        if has_spaces.any():
            print(f"Column '{col}' has leading or trailing spaces.")
            print(df.loc[has_spaces, col])
        else:
            print(f"Column '{col}' has no leading or trailing spaces.")
    else:
        print(f"Column '{col}' is not of string type, so no leading or trailing spaces to check.")
```

```python
#Create column for "Generation"
conditions = [
    (df['DOB'].between('1901-01-01', '1924-12-31')),
    (df['DOB'].between('1925-01-01', '1945-12-31')),
    (df['DOB'].between('1946-01-01', '1964-12-31')),
    (df['DOB'].between('1965-01-01', '1979-12-31')),
    (df['DOB'].between('1980-01-01', '1994-12-31')),
    (df['DOB'].between('1995-01-01', '2012-12-31')),
    (df['DOB'].between('2013-01-01', '2025-12-31'))
]

generation_labels = [
    "The Greatest Generation",
    "The Silent Generation",
    "Baby Boomers",
    "Generation X",
    "Millennials",
    "Generation Z",
    "Gen Alpha"
]

# Create the "Generation" column based on conditions
df['Generation'] = np.select(conditions, generation_labels, default='Unknown')

# Display the DataFrame
display(df)
```

```python
#Find date range for the datetime columns; DOB, DateofHire, DateofTermination, LastPerformanceReview_Date    

date_min_DOB = df['DOB'].min()
date_max_DOB = df['DOB'].max()

date_min_DateofHire = df['DateofHire'].min()
date_max_DateofHire = df['DateofHire'].max()

date_min_DateofTermination = df['DateofTermination'].min()
date_max_DateofTermination = df['DateofTermination'].max()

date_min_LastPerformanceReview_Date = df['LastPerformanceReview_Date'].min()
date_max_LastPerformanceReview_Date = df['LastPerformanceReview_Date'].max()

print(f"DOB Date range: {date_min_DOB} to {date_max_DOB}")
print(f"DateofHire Date range: {date_min_DateofHire} to {date_max_DateofHire}")
print(f"DateofTermination Date range: {date_min_DateofTermination} to {date_max_DateofTermination}")
print(f"LastPerformanceReview_Date Date range: {date_min_LastPerformanceReview_Date} to {date_max_LastPerformanceReview_Date}")
```
